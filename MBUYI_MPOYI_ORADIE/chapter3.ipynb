# 1.importation des packages

```python
# importer les Packages
import numpy as np
import pandas as pd
import tensorflow 
import seaborn as sns
import warnings 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

warnings.filterwarnings('ignore')
```

# 2.charger les données

```python
data = pd.read_csv('Data/DDoS_dataset.csv')
data.head()
```

# 3.Analyse de données 

```python
data.info()
```

```python
data['src'].unique()
```

```python
data['dst'].unique()
```

```python
data['Protocol'].unique()
```

```python
data.drop(columns='src', inplace=True, axis=1)
data.drop(columns='dst', inplace=True, axis=1)
data.head()
```

# 4.PRETRAITEMENT DE DONNEES

```python
encoder = LabelEncoder()
```

```python
data['Protocol'] = encoder.fit_transform(data['Protocol'])
data.head()
```

```python
data['Protocol'].unique()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
# gestion des valeurs manquantes
data['rx_kbps'].fillna(data['rx_kbps'].mean(), inplace=True)
data['tot_kbps'].fillna(data['tot_kbps'].mean(), inplace=True)
data.isnull().sum()
```

```python
data
```

```python
# 5. Diviser les données train et test (Normaliser les données)
```

```python
data.columns
```

```python
x= data.iloc[ : , :-1]
y= data.iloc[ : , -1]
```

```python
x
```

```python
y
```

```python
# splitting

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.23, random_state=40)
```

```python
print("x_train:", x_train.shape)
print("x_test:", x_test.shape)
print("y_train:", y_train.shape)
print("x_train:", x_test.shape)
```

# 6. construire ke shallow neural network

```python
model =  Sequential()
```

```python
# Ajout de la couche d'entree == couche cachee
model.add(Dense(10, input_shape=(20,), activation='relu'))
#model.add(Dense(16, input_shape=(9,), activation='relu'))
#model.add(Dense(32, activation='relu'))

# Ajout de la couche de sortie
model.add(Dense(1, activation='sigmoid'))

# Compiler le modele
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Affichage de l'architecture ou la structure du model
model.summary()
```

```python
# Entrainement du modele
history = model.fit(x_train, y_train, epochs=7, verbose=1, validation_data=(x_test, y_test))
```

# 7.visualser l'evoluion de l'entrainement du reseau neurone

```python
epoch_accuracy = history.history['accuracy']
epoch_loss = history.history['loss']

epoch_val_accuracy = history.history['val_accuracy']
epoch_val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(19, 10))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, color='green', label='Training Accuracy')
plt.plot(range(0, len(epoch_loss)), epoch_loss, color='red', label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Valeur')
plt.title('Accuracy Vs Loss')
plt.legend()
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(20, 7))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, color='green', label='Training Accuracy')
plt.plot(range(0, len(epoch_val_accuracy)), epoch_val_accuracy, color='red', label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training Accuracy Vs Test Accuracy')
plt.legend()
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(range(0, len(epoch_loss)), epoch_loss, color='green', label='Training Loss')
plt.plot(range(0, len(epoch_val_loss)), epoch_val_loss, color='red', label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Vs Test Loss')
plt.legend()
plt.plot()
plt.show()
```

###  Evaluer le modele

```python
model.evaluate(x_test, y_test)
```
